## 1. Introduction: Historical Aspects

**Radioactive decay** is the spontaneous change in the composition (charge Z, mass number A) or internal structure of unstable atomic nuclei through the emission of elementary particles, gamma quanta, and/or nuclear fragments. The process of radioactive decay is also referred to as **radioactivity** [[1](#bibliography)], [[2](#bibliography)].

At present, it is known next types of radioactive decay:

- **Alpha decay** — accompanied by the emission of alpha particles: the nucleus of a helium atom ($^{4}_{2}\text{He}$), consisting of 2 protons and 2 neutrons. Alpha radiation has low penetration power but high ionizing power.

- **Beta decay** — a decay process that causes the emission of beta particles (there are types of beta decay without beta particle emission, but beta decay is always accompanied by the emission of neutrinos or antineutrinos). Beta radiation has greater penetration than alpha particles but moderate ionizing power.

- **Gamma decay** — accompanied by the emission of gamma quanta, that has very high penetration power.

In addition to alpha ($\alpha$), beta ($\beta$), and gamma ($\gamma$) decays, other types of decay have been discovered, such as electron capture, positron decay (or β⁺ decay), cluster radiation, spontaneous fission, etc.

<i>Natural radioactivity</i> — the spontaneous decay of atomic nuclei found in nature.

<i>Artificial radioactivity</i> — the spontaneous decay of atomic nuclei obtained artificially through specific nuclear reactions.

Experiments have established that radioactive substances have no stable isotopes; all [chemical elements](https://en.wikipedia.org/wiki/List_of_elements_by_stability_of_isotopes) with atomic numbers greater than 82 are unstable.

<b>At the beginning of the study of radioactivity, key events can be recognized </b>:

In <b>1896</b>, <u>natural (spontaneous) radioactivity</u> was discovered by [Henri Becquerel](https://francearchives.gouv.fr/fr/pages_histoire/38898), 
> who found that uranium salts spontaneously emit invisible rays that darken photographic plates, even without exposure to sunlight. 
This discovery stimulated an active investigation into the phenomenon of radioactivity.

In <b>1898</b>, <u>the discovery of Polonium (Po) and Radium (Ra)</u> by [Marie and Pierre Curie](https://francearchives.gouv.fr/fr/pages_histoire/26287410).
> The new radioactive elements were isolated. It was demonstrated that radioactivity is a fundamental atomic property, not a result of a chemical reaction. 

In <b>1899–1900</b>, <u>[identification of different types of radiation](https://laradioactivite.com/categories/le-phenomene/rayonsalphabetagamma)</u>.
> Ernest Rutherford distinguished two types of radiation — alpha ($\alpha$) and beta ($\beta$).  
A third type of radiation - gamma rays (γ) - was discovered by Paul Villard.

<b>Early 20th century </b>, <u>theory of Radioactive Decay</u> and <u>Nuclear Model of the Atom </u>.
> were proposed by [Ernest Rutherford](https://laradioactivite.com/articles/le-phenomene/ernestrutherford). He demonstrated that atoms have a dense, positively charged nucleus with electrons in orbit. He also showed that $ \alpha $-particles are helium nuclei.  

<b>In 1928</b>, <u>the Alpha Decay theory</u> 
> was proposed by [George Gamow](https://odysseum.eduscol.education.fr/georges-gamow-ou-le-joyeux-refus-des-ideologies). The theory applies the new concepts of quantum mechanics to nuclear physics and explains alpha decay as a quantum tunneling process. 

These discoveries not only clarified the nature of alpha radiation but also provided a foundation for understanding the process of radioactive decay in atomic nuclei, paving the way for the development of nuclear physics and its applications in science and technology.

# 2. Modeling of Alpha-Particle Decay

## 2.1. Basic Principles and Definition of Radioactivity

As is well known, the atomic <b>nucleus</b> has a characteristic size on the order of $10^{-15}\text{m}$, carries a <b>positive charge</b>, and consists of <b>protons</b> and <b>neutrons</b>. The nucleons are bound within the nucleus by the <b>strong nuclear interaction</b>, which is short-ranged and acts at distances of about $10^{-15}\text{m}$.

Within this notebook, we consider unstable atoms that transform into atoms of other elements through the emission of $\alpha$ particles. The transformation can be expressed by the general nuclear reaction equation:

$$
{}^{A}_{Z}X \longrightarrow {}^{A-4}_{Z-2}Y + {}^{4}_{2}\alpha , 
$$


where ${}^{A}_{Z}X$ denotes the parent nucleus, ${}^{A-4}_{Z-2}Y$ the daughter nucleus, and $\alpha$ the emitted particle. 

Radioactivity is a random phenomenon that can be described using probability. For a macroscopic sample of radioactive matter containing a number $N$ of atoms, the number of decays during a time interval $t$ is expressed by the <b>law of radioactive decay</b>:

$$ N(t)=N_0 e^{−\Gamma t} \tag{1}$$

where: $N_0$ is the number of atoms at the initial time $t = 0$ and $Γ$ is the <b>decay constant</b>, representing the probability of decay per unit time, expressed in $s^{−1}$.

Another important characteristic is the <b>half-life</b> $t_{1/2}$, defined as the time required for half of the initially present radioactive atoms to decay:

$$ N(t_{1/2}) = \frac {N_0}{2} \quad \Rightarrow \quad  \frac{1}{2} = e^{− \Gamma t_{1/2}} \quad \Rightarrow \quad  t_{1/2} = \frac{\ln 2}{\Gamma} = \tau \ln 2, \tag{2} $$

where $\tau$ is the <b> mean lifetime</b> - the average lifetime of a radioactive particle before decay. 

## 2.2. Quantum approach to α decay

Within the nuclear interior, the strong force attraction between nucleons produces a deep well of potential energy, overcoming the strong electromagnetic repulsion of the protons. However, the strong nuclear force is an extremely short-range force; its influence disappears at the radius of the nucleus (Fig. 1).

$ \alpha $ decay cannot be explained by classical physics, since the alpha-particle with total energy $E_{\alpha}$ must overcome a high potential energy barrier, but it is well described by the quantum-mechanical process of tunnelling, in which an alpha particle has a finite probability of penetrating the barrier despite not having enough energy [[3](#bibliography)], [[4](#bibliography)]. 

The transmission probability is related to the decay constant $ \Gamma $ and the half-life of the radioactive nucleus $ \tau$. It can be defined as the ratio of the transmitted to incident probability currents, which are determined from the amplitudes of the wave function on either side of the potential barrier by solving the Schrödinger equation for a given potential:

$$T = \frac{J_\text{transmitted}}{J_\text{incident}} \tag{3}.$$



### 2.2.1.Aproximation Gamow

At first, the quantum-mechanical approach to describing $\alpha$ decay was proposed by G. Gamow in 1928 [[5](#bibliography)]. The potential energy of the  $\alpha$-particle, as a function of distance $r$ from the center of the nucleus, was represented schematically as a deep potential well inside the nucleus (due to strong nuclear forces), followed by a Coulomb term outside (arising from electrostatic repulsion), and tending to zero at large distances $r$:

$$ V_G(r) = 

\begin{cases}
V_m, &  r \le R, \\
\frac{2 (Z - 2)e^2}{r} ,  & r > R. \tag{4}
\end{cases}
$$

where $V_m$ is the depth of the potential well, which models the strong interaction, and $ R $ is the radius of the daughter nucleus, depending on the mass number $A$: 

$$ R = 1.07 \left(4^{1/3} + (A - 4)^{1/3} \right).  \tag{5} $$

To determine the transmission coefficient of the $\alpha$ particle through the potential barrier, a wide potential barrier of arbitrary shape can be approximated as a sequence of $n$ thin rectangular potential barriers with heights $V_r$ and constant width $dr$ provided that the wave function continuity conditions are satisfied (see Fig.2 [[4](#bibliography)]). In this case, the <b>tunneling probability</b> through the entire barrier is

$$
T \simeq \prod_{i=1}^{n} 
    \exp\!\left[-\frac{2}{\hbar}\,\sqrt{2m\,(V_i - E)}\, dr \,\right] \tag{6} 

$$

Within the framework of the proposed model of nuclear potential, Gamow proposed a quantitative relationship between the half-life and the energy of the emission particle using first principles.

By considering the time interval between two successive attempts to cross the barrier:

$$ \tau_F = \frac{v}{2R}\; \text{with}\; v = \sqrt{\frac{2E}{m}} ,$$

where $v$ is the particle’s velocity inside the nucleus, the half-life can be expressed as:
$$ t_{1/2} = \frac{\tau_F \ln 2}{T}.$$

As shown in [[4](#bibliography)], this approach yields good agreement with experimental data.

In the limit $dr \to 0$ expression (6) leads to the WKB (Wentzel–Kramers–Brillouin) approximation [[6](#bibliography)], providing a practical method for calculating the tunneling coefficient of a potential barrier. This approach was first applied by Gamow in 1928 to quantitatively model alpha decay.

### 2.2.2. Aproximation WKB

In general, the WKB (Wentzel–Kramers–Brillouin) approximation is a technique for finding approximate solutions to linear differential equations with spatially varying coefficients. 

In quantum mechanics, it is typically used to solve the one-dimensional Schrödinger equation for a system where the potential energy is not uniform but changes slowly compared to the particle’s wavelength.

The time-independent Schrödinger equation can be written as

$$
-\frac{\hbar^2}{2m} \frac{d^2 \psi(r)}{dr^2} + V(r)\psi(r) = E\psi(r). \tag{7}
$$

In this method, the wave function is expressed as an exponential function, with either the amplitude or the phase assumed to vary slowly. The WKB approximation is essentially a semi-classical method, in which the motion of the particle is treated using classical concepts (such as classical momentum), while quantum effects are incorporated through the wave nature of particles.

The general WKB solution for classically allowed region $E > V(r)$, when $k(r) = \frac{\sqrt{2m(E-V(r))}}{\hbar} $ is given by

$$
\psi(r) = \frac{A}{\sqrt{k(r)}} \exp\left(i \int^r k(r)\,dr\right) + \frac{B}{\sqrt{k(r)}} \exp\left( -i \int^r k(r)\,dr\right), \tag{8}
$$

In the case of the classically forbidden region $E < V(r)$, where $q(r) = \frac{\sqrt{2m(V(r)-E)}}{\hbar} $ is given by

$$
\psi(r) = \frac{C}{\sqrt{q(r)}} \exp\left(-\int^r q(r)\,dr\right) + \frac{D}{\sqrt{q(r)}} \exp\left( \int^r q(r)\,dr\right), \tag{9}
$$

where $A$, $B$, $C$, $D$ and $F$ are constants determined by the boundary conditions.

For potentiel prsented in (4) WKB appriximation give the next solution for wave functiom in different region:

$$ 
\begin{cases}
\psi_I(r) = Ae^{ik_1(r-r_1)} + Be^{-ik_1(r-r_1)}, &  r < r_1, \\
\psi_{II}(r) = \frac{C}{\sqrt{q(r)}} e^{ -\int q(r) dr} + \frac{D}{\sqrt{q(r)}} e^{\int q(r) dr } ,  & r_1 <= r <= r_2 \tag{10}\\
\psi_{III}(r) = Fe^{ik_3(r-r_2)}, & r \gg r_2
\end{cases}
$$

where $ k_1 = \frac{\sqrt{2m(E-V_m)}}{\hbar} $,  $ k_3 = \frac{\sqrt{2mE}}{\hbar} $ and $q(r) = \frac{\sqrt{2m(V(r) - E)}}{\hbar}.$ 

To find the coefficients A, B, C, D and F we must have continuity of the wavefunction and its derivative:

$$ \psi_I(r_1) = \psi_{II}(r_1), \\ \psi_I'(r_1) = \psi_{II}'(r_1) .$$

Assuming that $\psi_{III}$ is valid for region $r > r_2$:
$$ \psi_{II}(r_2) = \psi_{III}(r_2),\\ \psi_{II}'(r_2) = \psi_{III}'(r_2) .$$


Wavefunction continuity at $r = r_1$:  $$ A + B  = \frac{C + D}{\sqrt{q(r_1)}} .$$

Derivative continuity at $r = r_1$ and under the condition weak change of the potential:
$$ ik_1 (A - B)  = (-C + D) \sqrt {q(r_1)} .$$

Wavefunction continuity at $r = r_2$:  $$ \frac{C e^{ -\int_{r_1}^{r_2} q(r) dr} + D e^{ \int_{r_1}^{r_2} q(r) dr}}{\sqrt{q(r_2)}} = F .$$

Derivative continuity at $r = r_1$ and under the condition weak change of the potential:
$$ (-C e^{ -\int_{r_1}^{r_2} q(r) dr} + D e^{ \int_{r_1}^{r_2} q(r) dr}) \sqrt {q(r_2)} =  ik_3 F .$$


To solve the equation system we assume the amplitude of incoming wave $A = 1$. THe wave function profile is presente in the next section.


Transmission coefficient:

$$ T = \frac{k_3|F|^2}{k_1|A|^2} = \frac{k_3}{k_1}|C|^2 e^{-2 \int_{r_1}^{r_2}q(r)dr} = \frac{k_3}{k_1} e^{ - \frac{2}{\hbar} \int_{r_1}^{r_2} \sqrt {2 m [V(r) - E] dr} } .\tag{11}$$

## 2.3. Modeling of Alpha Decay Using the WKB Method

The total potential used for modeling $\alpha$-decay can be expressed as the sum of nuclear potentiel (described by Woods-Saxon potentiel), the centrifugal potential and Coulomb potential [[4](#bibliography)], [[6](#bibliography)].

$$ V(r) = V_C(r) + V_l(r) + V_{WS} . \tag{3}$$

The Coulomb potential given by

$$ V_C(r) = 

\begin{cases}
\frac{2(Z - 2) e^2}{2 R_0^3} \left[3 R_0^2 - r^2\right] , &  r \le R_0, \\
\frac{2 (Z - 2)e^2}{r} ,  & r > R_0; \tag{4}
\end{cases}
$$

the Woods-Saxon nuclear potentiel is calculated as

$$ V_{WS} = - \frac {V0} {(1 + \exp {(r - R0)/a})}; \tag{5}$$

the centrifugal potentiel is determined as

$$ V_l =  \frac{\hbar^2 l (l + 1)} {2 \mu r^2}, \tag{6}$$

where $Z$ is the atomic number, $R_0$ is the nuclear radius, $V_0$ - depth of nuclear potential well, $a$ - diffuseness parameter, $l$ and $\mu$ - the $\alpha$ particle angular momentum and the reduced mass of the system formed by the alpha particle and the daughter nucleus, respectively.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import quad
from scipy.optimize import brentq

# -----------------------------------------------------------------------------
# 1. Global Definitions and Parameters (in atomic units)
# -----------------------------------------------------------------------------

ALPHA_MASS_AU = 7294.0     # alpha particle mass in me
# AU_TIME_TO_S = 1  # 1 a.u. of time -> seconds
AU_TIME_TO_S = 2.418884e-17  # 1 a.u. of time -> seconds: Atomic unit of time to seconds conversion

# -----------------------------
# Input Parameters
# -----------------------------
element_data = {
    #Polonium
    "Po-210": {"Z": 84, "E_MeV": 5.3, "V0_MeV": 62, "R0_fm": 7.5, "a_fm": 0.69,"l_orbital": 0, "r_min_fm": 0.5, "r_max": 5e-3}, #second model  

    #Radium
    "Ra-226": {"Z": 88, "E_MeV": 4.78, "V0_MeV": 62, "R0_fm": 8.5, "a_fm": 0.65, "l_orbital": 0, "r_min_fm": 0.5, "r_max": 5e-3},
    
    #Uranium
    "U-238": {"Z": 92, "E_MeV": 4.56, "V0_MeV": 66, "R0_fm": 8.7, "a_fm": 0.65, "l_orbital": 0, "r_min_fm": 0.5, "r_max": 5e-3},
    
    #Thorium
    "Th-232": {"Z": 90, "E_MeV": 4.082, "V0_MeV": 50, "R0_fm": 8.7, "a_fm": 0.65, "l_orbital": 0, "r_min_fm": 0.5, "r_max": 5e-3},

    #Plutonium
    "Pu-238": {"Z": 94, "E_MeV": 5.593, "V0_MeV": 63, "R0_fm": 8.2, "a_fm": 0.7, "l_orbital": 0, "r_min_fm": 0.5, "r_max": 5e-3},


}

# Unit conversions
def MeV_to_Hartree(MeV): return MeV * 36749.7
def fm_to_Bohr(fm): return fm * 1.8897e-5

# Choose element : Polonium - "Po-210", 
#                  Radium   - "Ra-226"
#                  Uranium  - "U-238", 
#                  Thorium  - "Th-232"

element = "Po-210"

# Load parameters from dictionary
params = element_data[element]

#Apply to your global variables
Z_parent    = params["Z"]                       # Parent nucleus charge
E_hartree   = MeV_to_Hartree(params["E_MeV"])   # Alpha particle energy
V0_hartree  = MeV_to_Hartree(params["V0_MeV"])  # Depth of nuclear potential well, usually in the range of 40 to 60 MeV for heavy nuclei.
R0_bohr     = fm_to_Bohr(params["R0_fm"])       # Nuclear radius in femtometers (fm). R0≈1.2×A^(1/3) fm => where A is the mass number. Need to be adjusted.
a_bohr      = fm_to_Bohr(params["a_fm"])        # diffuseness (Bohr): Physically, it's the distance over which the nuclear potential drops from ~90% of V0 to ~10%.


l_orbital   = params["l_orbital"]               # Orbital angular momentum of alpha particle
r_min       = fm_to_Bohr (params["r_min_fm"])                  # inner bohr-radius
# r_min       = params["r_min_fm"]                # inner bohr-radius
r_max       = R0_bohr * 30.0                    # outer radius ~ 30 * R0 (Bohr)              # outer bohr-radius
N           = 2000                              # number of points for scanning turning points

# Optional: print to confirm
print(f"Loaded parameters for {element}:")
print(f"  Z_parent = {Z_parent}, E_hartree = {E_hartree:.2e} Ha, V0_hartree = {V0_hartree:.2e} Ha, R0_bohr = {R0_bohr:.2e} Bohr, a_bohr = {a_bohr:.2e} Bohr, r_min = {r_min}, r_max = {r_max}")
```

```python
# -----------------------------------------------------------------------------
# 2. Potential: Woods-Saxon + Coulomb in a.u.
# -----------------------------------------------------------------------------

def V_WS_Coulomb_au(r, V0, R0, a, Z_nuc, Zalpha=2, mass_alpha=ALPHA_MASS_AU, l=0):
    """
    Woods-Saxon + Coulomb + optional centrifugal term in atomic units.
    """

    e = 1
    if r < 1e-12:
        r = 1e-12
 
    # Woods-Saxon
    Vws = - V0 / (1.0 + np.exp((r - R0)/a))

    # Coulomb
    if r <= R0:
        Vcoul = (2 * (Z_nuc - 2) * e**2 * (3*R0**2 - r**2)) / (2 * R0**3)
        # Vw  = V1 * np.exp(-r/sigma)

    elif r > R0:
        Vcoul = 2 * (Z_nuc - 2) * e**2 / r
        # Vw = 0

    # Centrifugal
    Vcentrif = l * (l + 1) / (2.0 * mass_alpha * r * r)

    return Vws + Vcoul + Vcentrif


# Potential Wrapper
def potential_au(r):
    return V_WS_Coulomb_au(r,
                           V0=V0_hartree,
                           R0=R0_bohr,
                           a=a_bohr,
                           Z_nuc=Z_parent,
                           Zalpha=2,
                           mass_alpha=ALPHA_MASS_AU,
                           l=l_orbital)

# -----------------------------------------------------------------------------
# 3. WKB and Classical Integrand Functions
# -----------------------------------------------------------------------------
def forbidden_integrand(r, E, m_alpha, pot_func):
    """
    Calculates the square root for the WKB integral
    sqrt(2*m_alpha*(V(r)-E)) if V(r)>E, else 0
    """
    val = pot_func(r) - E
    return np.sqrt(2*m_alpha*val) if val > 0 else 0.0

def classical_integrand(r, E, m_alpha, pot_func):
    """
    Calculates inverse velocity in the classical region
    1 / v(r) = sqrt(m_alpha/(2*(E - V(r)))) if E>V(r), else 0
    """
    val = E - pot_func(r)
    return np.sqrt(m_alpha/(2*val)) if val > 0 else 0.0

# -----------------------------------------------------------------------------
# 4. Helper Functions
# -----------------------------------------------------------------------------
def find_turning_points(r_min, r_max, E, pot_func, N=2000):
    r_vals = np.linspace(r_min, r_max, N)
    diff = E - np.array([pot_func(rv) for rv in r_vals])
    
    turning_pts = []
    for i in range(len(r_vals)-1):
        if diff[i]*diff[i+1] < 0:
            root = brentq(lambda rr: E - pot_func(rr), r_vals[i], r_vals[i+1])
            turning_pts.append(root)
    turning_pts.sort()
    return turning_pts

def detect_barrier_hill(E, turning_points, r_min, r_max, pot_func):
    """
    Looks for the pattern (+)->(-)->(+) in E-V(r)
    Returns the last such (r_in, r_out) if found
    """
    boundaries = [r_min] + turning_points + [r_max]
    signs = []
    for i in range(len(boundaries)-1):
        mid = 0.5*(boundaries[i] + boundaries[i+1])
        val_mid = E - pot_func(mid)
        s = 1 if val_mid > 0 else -1
        signs.append(s)
    
    hill_intervals = []
    for i in range(len(signs)-2):
        if signs[i] == 1 and signs[i+1] == -1 and signs[i+2] == 1:
            hill_start = boundaries[i+1]
            hill_end   = boundaries[i+2]
            hill_intervals.append((hill_start, hill_end))
    
    if hill_intervals:
        return hill_intervals[-1]
    return None

def compute_WKB_barrier_integral(r_in, r_out, E, m_alpha, pot_func):
    val, _ = quad(lambda rr: forbidden_integrand(rr, E, m_alpha, pot_func),
                  r_in, r_out, epsabs=1e-12, epsrel=1e-12, limit=1000)
    return val

def find_well_region(E, turning_points, r_min, r_max, pot_func):
    """
    Finds the first sub-interval [r_a, r_b] where E>V(r)
    """
    boundaries = [r_min] + turning_points + [r_max]
    for i in range(len(boundaries)-1):
        left = boundaries[i]
        right = boundaries[i+1]
        mid = 0.5*(left + right)
        if (E - pot_func(mid)) > 0:
            return (left, right)
    return None

def compute_classical_period(r_a, r_b, E, m_alpha, pot_func):
    delta = 1e-6 * (r_b - r_a)

    single_trip, _ = quad(lambda rr: classical_integrand(rr, E, m_alpha, pot_func),
                           r_a + delta, r_b - delta, epsabs=1e-12, epsrel=1e-12, limit=1000)
    return 2.0 * single_trip

# -----------------------------------------------------------------------------
# 5. Troubleshooting
# -----------------------------------------------------------------------------

# Evaluates the potential on a grid to see if there's a barrier above E
r_dense = np.linspace(r_min, r_max, 2000)
V_dense = np.array([potential_au(rv) for rv in r_dense])

# Prints min/max of potential
V_min = np.min(V_dense)
V_max = np.max(V_dense)
print("Troubleshooting Info:")
print(f"  r_min = {r_min}, r_max = {r_max}")
print(f"  V_min over this range = {V_min:.4e} Hartree")
print(f"  V_max over this range = {V_max:.4e} Hartree")
print(f"  E (alpha energy)      = {E_hartree:.4e} Hartree")

# Checks if E is bigger than V_max or smaller than V_min
if E_hartree > V_max:
    print("E is ABOVE the maximum potential => no barrier (region with V>E)")
if E_hartree < V_min:
    print("E is BELOW the entire potential => might only have 1 crossing or none")

# Sample a few points in E - V(r) to determine sign
sample_indices = [0, len(r_dense)//2, len(r_dense)-1]
print("\nSample checks of sign(E - V(r)) at a few points:")
for idx in sample_indices:
    rr = r_dense[idx]
    val = E_hartree - V_dense[idx]
    sign_str = "+" if val>0 else "-"
    print(f"  r={rr:.4e}, V(r)={V_dense[idx]:.4e}, E-V={val:.4e} ({sign_str})")

# -----------------------------------------------------------------------------
# 6. Main Calculation
# -----------------------------------------------------------------------------
print("\n-- Attempting to detect barrier & compute WKB --")

turning_pts = find_turning_points(r_min, r_max, E_hartree, potential_au, N=N)
print(f"Turning points found (Bohr radii): {turning_pts}")

barrier = detect_barrier_hill(E_hartree, turning_pts, r_min, r_max, potential_au)

if barrier is None:
    raise ValueError("No barrier (hill) detected.")
r_in, r_out = barrier
print(f"Barrier region: r_in={r_in:.4e}, r_out={r_out:.4e}")

barrier_integral = compute_WKB_barrier_integral(r_in, r_out, E_hartree, ALPHA_MASS_AU, potential_au)
P_tunnel = np.exp(-2.0 * barrier_integral)

well_region = find_well_region(E_hartree, turning_pts, r_min, r_max, potential_au)
if well_region is None:
    raise ValueError("No well region found (E < V everywhere?).")
r_a, r_b = well_region
T_au = compute_classical_period(r_a, r_b, E_hartree, ALPHA_MASS_AU, potential_au)
T_s = T_au * AU_TIME_TO_S  # 1 a.u. of time -> seconds

Gamma_s = (1.0 / T_s) * P_tunnel
t_half_s = np.log(2) / Gamma_s

print("\n-- Results --")
print(f"WKB barrier integral = {barrier_integral:.6e}")
print(f"Tunneling probability = {P_tunnel:.3e}")
print(f"Well region: r_a={r_a:.4e}, r_b={r_b:.4e}")
print(f"Classical period T_au = {T_au:.4e} => T_s={T_s:.4e} s")
print(f"Decay rate Gamma = {Gamma_s:.4e} s^-1")
print(f"Half-life t_1/2 = {t_half_s:.4e} s")

# -----------------------------------------------------------------------------
# 8. Plot
# -----------------------------------------------------------------------------
plt.figure(figsize=(8,6))
plt.plot(r_dense, V_dense, label='V(r)')
plt.axhline(E_hartree, color='r', linestyle='--', label=f'E={E_hartree:.2e}')
plt.axvline(r_in, color='k', linestyle='--', label='r1 (inner turning point)')
plt.axvline(r_out, color='m', linestyle='--', label='r2 (outer turning point)')
plt.xlabel('r (Bohr)')
plt.ylabel('V(r) (Hartree)')
plt.title('Nuclear Potential in Atomic Units')
plt.grid(True)
plt.legend()
plt.xlim(1e-6, 1e-3)
plt.ylim(-3e6, 2e6)
plt.show()
```

# 3. Applications of Alpha Decay

Medicine, energy generation, cancer treatment, detectors, environmental monitoring, and materials science are all areas where alpha decay finds applications. 

In this notebook, we will focus on the operating principle of Radioisotope Thermoelectric Generators (RTGs), where alpha decay plays a central role. In an RTG, the energy released by the spontaneous emission of alpha particles from radioactive isotopes is converted into heat. When an alpha-emitting isotope, such as Plutonium-238, undergoes decay, each emitted alpha particle carries significant kinetic energy, which is absorbed by the surrounding material and transformed into thermal energy. This heat is then converted into electricity using thermocouples through the Seebeck effect. Unlike nuclear fission reactors, alpha decay does not produce neutrons or sustain a chain reaction; instead, it provides a continuous and reliable heat source over long periods. RTGs leverage this property to supply power in environments where conventional energy sources are impractical, such as deep-space missions or remote installations.



RTGs are particularly suitable as energy sources for autonomous systems located far from conventional power grids, where a few tens to hundreds of watts are needed over very long durations—far longer than typical batteries or fuel cells can provide.

For example, RTGs have been used to power spacecraft such as the Voyager probes, the Mars Curiosity rover, and the New Horizons mission to Pluto, providing reliable electricity for instruments and communication systems in deep space where solar power is insufficient.

### Problem:

A Radioisotope Thermoelectric Generator contains 1 kg of Plutonium-238 ($^{238} \text{Pu}$). The energy released during the alpha decay of a single nucleus is 5.593 MeV. The generator’s thermocouples convert the thermal energy into electricity with an efficiency of 6%.

- Calculate the thermal power produced by 1 kg of Plutonium-238.

- Determine the electrical power output of the RTG.

- If the RTG is used to power electronics on a spacecraft, calculate how its power output will change after 10 years.

Parameters used:

- Mass of Plutonium-238: $m$ = 1 kg

- Energy per alpha particle: $E_{\alpha}$ = 5.593 MeV

- Thermocouple efficiency: $\eta =6\% = 0.06$

- Use the code below to calculate the half-life $T_{1/2}$ = 


Compare the efficiency of the RTG when the energy source is another radioactive material (Th-232, U-238, Po-210, etc.). What amount of uranium would be required to achieve similar generator performance?


<b> Step 1: Calculate the number of atoms in 1 kg of Plutonium-238 </b>

Molar mass of Pu-238: 
$$M = 0.238 \text{kg/mol}$$

Number of atoms: 
$$ N = \frac{m}{M} N_A $$


<b> Step 2: Calculate the decay constant </b>

$$ \lambda = \frac{\ln{2}}{T_{1/2}} $$

<b> Step 3: Radioactive decay law: </b> Number of undecayed nuclei at time $t$:

$$N(t) = N_0 e^{-\lambda t}$$

<b>Step 4: Calculate the thermal power </b> (energy per decay × decays per second): Each alpha decay releases (on average) energy $E_{\alpha}$ carried mainly by the alpha particle. That kinetic energy stops locally and becomes heat with efficiency very close to 1.

at $t = 0$

$$P_{heat} = \lambda \cdot N_0 \cdot E_{\alpha}$$

with time dependence:

$$P_{heat}(t) = \lambda \cdot N_0 e^{-\lambda t}  \cdot E_{\alpha}$$

<b>Step 5: Electrical power, if the thermoelectric efficiency is $\eta$:</b>

$$ P_{el} = \eta \cdot P_{heat} $$

```python
import math
import matplotlib.pyplot as plt
import numpy as np

m = 1.0  # kg of Plutonium-238
M = 0.238  # kg/mol
N_A = 6.022e23  # Avogadro's number
T_half_years = 87.7  # half-life in years
E_alpha_MeV = 5.593  # energy per alpha particle in MeV
eta = 0.06  # thermocouple efficiency

seconds_per_year = 365 * 24 * 3600
T_half_sec = T_half_years * seconds_per_year

E_alpha_J = E_alpha_MeV * 1.602e-13

# Initial number of atoms
N0 = (m / M) * N_A

# Decay constant
lambda_decay = math.log(2) / T_half_sec

# Time array (0 to 200 years)
years = np.linspace(0, 200, 500)
seconds = years * seconds_per_year

# Number of atoms over time
N_t = N0 * np.exp(-lambda_decay * seconds)

# Thermal and electrical power over time
P_heat_t = lambda_decay * N_t * E_alpha_J
P_elec_t = eta * P_heat_t

# Plot results
plt.figure(figsize=(12,5))

# Plot number of atoms vs years
plt.subplot(1,2,1)
plt.plot(years, N_t/1e24, label="Remaining atoms (×10^24)")
plt.xlabel("Time (years)")
plt.ylabel("Number of atoms (×10^24)")
plt.title("Decay of Pu-238")
plt.grid(True)
# plt.legend()

# Plot electrical power vs years
plt.subplot(1,2,2)
plt.plot(years, P_elec_t, color="orange", label="Electrical Power (W)")
plt.xlabel("Time (years)")
plt.ylabel("Electrical Power (W)")
plt.title("RTG Electrical Power Output")
plt.grid(True)
# plt.legend()

plt.tight_layout()
plt.show()
```
